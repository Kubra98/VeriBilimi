```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
yaslistesi=[10,40,60,75,25,20,45,30,55,35]
kilolistesi=[25,50,55,60,65,70,55,60,65,55]
numpyyaslistesi= np.array(yaslistesi)
numpykilolistesi= np.array(kilolistesi)
plt.plot(numpykilolistesi,numpyyaslistesi, "r--")# r red anlamı veriyor kırmızı çizgi ile gösteriyor
plt.xlabel("kilo")
plt.ylabel("Yaş")
plt.title("Kilonun yaşa göre değişim grafiği")
```

```python
plt.subplot(1,2,1) # 1 sırada 2 kolon olacak 1. grafik şuan diyip
plt.plot(numpykilolistesi,numpyyaslistesi, "r*-")
plt.subplot(1,2,2)
plt.plot(numpykilolistesi,numpyyaslistesi, "b*-") #en sondaki 2 de 2. grafik şuan 
```

```python
figure= plt.figure()# boş bir  dikdörtgen figure açar bize sadece
figureAxes= figure.add_axes([0.2,0.2,0.9,0.9])#son ikisi width ve height
figureAxes.plot(numpykilolistesi,numpyyaslistesi, "b")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik başlığı")
```

```python
figure2= plt.figure()
eksen1= figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2= figure2.add_axes([0.2,0.3,0.3,0.3])
eksen1.plot(numpykilolistesi,numpyyaslistesi,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana grafik başlık")

eksen2.plot(numpykilolistesi,numpyyaslistesi,"g")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
plt.subplots()
```

```python
(benimfigure, benimeksen)=plt.subplots(nrows=1,ncols=2)
benimeksen.plot(numpykilolistesi,numpyyaslistesi,"g")


```

```python
for eksen in benimeksen:
    eksen.plot(numpykilolistesi,numpyyaslistesi,"r")
    eksen.set_xlabel("X ekseni")
plt.tight_layout()# 2  grafik arasına boşluk koyar sıkışık olduğunda
benimfigure
```

```python
yenifigure=plt.figure(figsize=(6,6),dpi=80) # normalde içi boştu size verdik figure e vermezsek otomatik yapar
#dpi=100 de daha kaliteli gözükmesini sağladı rakam değişebilir
yenieksen= yenifigure.add_axes([0.1,0.1,0.9,0.9])
yenieksen.plot(numpykilolistesi,numpykilolistesi **2, label="numpykilolistesi ** 2")
yenieksen.plot(numpykilolistesi,numpykilolistesi ** 3, label="numpykilolistesi ** 3")
yenieksen.legend(loc=2)#locasyon tanımlar bize bunu değiştirebiliriz.
#parantes içine 1 2 3 4 5 6 denilebilir
```
